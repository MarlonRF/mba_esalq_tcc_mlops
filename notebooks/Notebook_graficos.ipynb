# Pairplot com KDE na diagonal

```python
# PAIRPLOT COM ENVELOPES KDE
import textwrap  # biblioteca para "quebrar linha" em strings (til em rtulos longos)
import seaborn as sns  # biblioteca de visualizao (pairplot, kdeplot)
import matplotlib.pyplot as plt  # controle de figura, layout e exibio


def quebrar_texto(rotulo, largura=12):
    # Converte o rtulo para string e quebra em mltiplas linhas com no mximo "largura" caracteres por linha
    return textwrap.fill(str(rotulo), largura)


def remover_todas_legendas(grafico):
    # Percorre todos os eixos do grid do seaborn (cada subplot)
    for eixo in grafico.axes.flatten():
        # Se por algum motivo o eixo no existir, pula
        if eixo is None:
            continue
        # Obtm a legenda do eixo (se existir)
        legenda = eixo.get_legend()
        # Se existir legenda, remove (isso evita legendas duplicadas)
        if legenda is not None:
            legenda.remove()


def adicionar_legenda_unica_fora(
    grafico,
    titulo=None,                 # ttulo opcional da legenda (ex.: "ms-ano")
    local="center left",         # posio da legenda (em relao ao anchor)
    ancora=(1.02, 0.5),          # ncora fora da figura (x > 1 joga para direita)
    margem_direita=0.85,         # reserva espao no lado direito para a legenda
    fonte=16,                    # tamanho da fonte dos itens da legenda
    fonte_titulo=18              # tamanho da fonte do ttulo da legenda (se houver)
):
    # Pega os "handles" e "labels" (smbolos e textos) a partir do 1 eixo do grid
    handles, labels = grafico.axes[0, 0].get_legend_handles_labels()

    # Cria uma legenda nica na figura inteira (no em cada eixo)
    grafico.fig.legend(
        handles,                 # smbolos (marcadores)
        labels,                  # nomes das classes/grupos
        title=titulo,            # ttulo da legenda
        loc=local,               # localizao relativa
        bbox_to_anchor=ancora,   # coordenada (x,y) para ancorar a legenda
        frameon=False,           # remove a "caixa" em volta da legenda
        fontsize=fonte,          # tamanho da fonte dos itens
        title_fontsize=fonte_titulo  # tamanho da fonte do ttulo
    )

    # Ajusta o espao til dos subplots, deixando a faixa da direita livre para a legenda
    grafico.fig.subplots_adjust(right=margem_direita)


def sobrepor_contornos_kde(
    grafico,
    dados,
    variaveis,
    coluna_grupo,
    paleta=None,           # dicionrio de cores por grupo (opcional)
    niveis=6,              # quantidade de contornos (mais = mais "anis")
    limiar=0.05,           # corta caudas do KDE (evita contornos muito espalhados)
    ajuste_bw=1.2,         # suavizao do KDE (maior = mais suave)
    largura_linha=1.0      # espessura dos contornos
):
    # Percorre a matriz de subplots do pairplot por posio (i, j)
    for i, var_y in enumerate(variaveis):
        # Percorre as variveis do eixo X
        for j, var_x in enumerate(variaveis):
            # Seleciona o eixo (subplot) correspondente (linha i, coluna j)
            eixo = grafico.axes[i, j]
            # Se no existir eixo (casos raros) ou se for diagonal (i == j), pula
            if eixo is None or i == j:
                continue

            # Desenha o KDE 2D por grupo (hue) como contornos sobre o scatter j existente
            sns.kdeplot(
                data=dados,              # DataFrame com os dados
                x=var_x,                 # varivel no eixo X
                y=var_y,                 # varivel no eixo Y
                hue=coluna_grupo,        # coluna que define os grupos (cores)
                levels=niveis,           # nmero de nveis/contornos
                fill=False,              # no preenche, s desenha contornos
                thresh=limiar,           # limiar para ignorar densidade baixa (caudas)
                bw_adjust=ajuste_bw,     # ajuste de banda (suavizao)
                common_norm=False,       # normaliza KDE separadamente por grupo (melhor p/ envelopes)
                palette=paleta,          # cores fixas por grupo (se voc passou)
                linewidths=largura_linha,# espessura das linhas de contorno
                ax=eixo,                 # eixo alvo onde desenhar
                legend=False             # impede o KDE de criar legenda (evita duplicao)
            )


def quebrar_ticks(grafico, largura=12):
    # Percorre todos os eixos do grid
    for eixo in grafico.axes.flatten():
        # Se o eixo no existir, pula
        if eixo is None:
            continue
        # Atualiza os rtulos do eixo X quebrando texto em vrias linhas
        eixo.set_xticklabels([quebrar_texto(t.get_text(), largura) for t in eixo.get_xticklabels()])
        # Atualiza os rtulos do eixo Y quebrando texto em vrias linhas
        eixo.set_yticklabels([quebrar_texto(t.get_text(), largura) for t in eixo.get_yticklabels()])


def pairplot_com_envelopes_kde(
    dados,
    variaveis,
    coluna_grupo,
    marcadores=None,              # dict: {grupo: marcador} ex.: {"08-2015":"o", ...}
    paleta=None,                  # dict: {grupo: cor} ex.: {"08-2015":"#...", ...}
    tamanho_ponto=20,             # tamanho do ponto (scatter)
    alpha_ponto=0.70,             # transparncia do ponto
    cor_borda="white",            # cor da borda do ponto
    espessura_borda=0.5,          # espessura da borda do ponto
    diag_preencher=True,          # preencher KDE na diagonal
    diag_alpha=0.25,              # transparncia do preenchimento na diagonal
    kde_niveis=6,                 # nmero de contornos KDE fora da diagonal
    kde_limiar=0.05,              # corte de caudas do KDE
    kde_ajuste_bw=1.2,            # suavizao do KDE
    kde_largura_linha=1.0,        # espessura do contorno KDE
    largura_quebra_tick=12,       # largura para quebra de texto dos ticks
    titulo_legenda=None           # ttulo da legenda (opcional)
):
    # Cria o pairplot: disperso fora da diagonal e KDE na diagonal
    grafico = sns.pairplot(
        dados,                     # DataFrame com os dados
        vars=variaveis,            # lista de variveis numricas a plotar
        hue=coluna_grupo,          # coluna categrica que define grupos (cores/smbolos)
        markers=marcadores,        # marcadores por grupo (se fornecido)
        palette=paleta,            # cores por grupo (se fornecido)
        diag_kind="kde",           # diagonal com KDE (em vez de histograma)
        plot_kws={                 # parmetros do scatter
            "s": tamanho_ponto,    # tamanho dos pontos
            "alpha": alpha_ponto,  # transparncia
            "edgecolor": cor_borda,# cor da borda
            "linewidth": espessura_borda,  # espessura da borda
        },
        diag_kws={                 # parmetros do KDE na diagonal
            "fill": diag_preencher,# preenche KDE
            "alpha": diag_alpha,   # transparncia do preenchimento
        },
    )

    # Sobrepe contornos KDE 2D (envelopes) nos subplots fora da diagonal
    sobrepor_contornos_kde(
        grafico=grafico,           # PairGrid retornado pelo pairplot
        dados=dados,               # DataFrame
        variaveis=variaveis,       # lista de variveis
        coluna_grupo=coluna_grupo, # coluna de grupos
        paleta=paleta,             # paleta consistente com o scatter
        niveis=kde_niveis,         # n de contornos
        limiar=kde_limiar,         # corte de caudas
        ajuste_bw=kde_ajuste_bw,   # suavizao
        largura_linha=kde_largura_linha  # espessura
    )

    # Se voc quiser quebrar rtulos dos ticks (opcional)
    if largura_quebra_tick is not None:
        quebrar_ticks(grafico, largura=largura_quebra_tick)

    # Remove qualquer legenda que tenha surgido em eixos individuais
    remover_todas_legendas(grafico)

    # Cria uma legenda nica, fora do grfico
    adicionar_legenda_unica_fora(
        grafico,
        titulo=titulo_legenda
    )

    # Ajusta layout para no cortar elementos (principalmente ticks/legenda)
    plt.tight_layout()

    # Retorna o objeto do seaborn caso voc queira reutilizar (salvar, ajustar, etc.)
    return grafico


```

```python
# -----------------------------
# USO (exemplo)
# -----------------------------
variaveis = ["tmedia", "tu", "ur"]      # variveis que iro compor o pairplot
coluna_grupo = "mes-ano"               # coluna do DataFrame que define os agrupamentos

marcadores = {                         # marcador por grupo (tem que bater com os valores em df[coluna_grupo])
    "08-2015": "o",                    # crculo
    "01-2016": "s",                    # quadrado
    "07-2016": "D",                    # losango
}

paleta = {                             # cor por grupo (opcional)
    "08-2015": "#19850F",
    "01-2016": "#1C1A99",
    "07-2016": "#FF8C00",
}

grafico = pairplot_com_envelopes_kde(
    dados=df,                          # seu DataFrame
    variaveis=variaveis,               # colunas numricas
    coluna_grupo=coluna_grupo,         # coluna categrica
    marcadores=marcadores,             # markers por grupo
    paleta=paleta,                     # cores por grupo
    tamanho_ponto=20,                  # tamanho dos pontos
    alpha_ponto=0.70,                  # transparncia dos pontos
    cor_borda="white",                 # borda branca
    espessura_borda=0.5,               # espessura da borda
    diag_preencher=True,               # KDE diagonal preenchido
    diag_alpha=0.25,                   # transparncia do KDE na diagonal
    kde_niveis=6,                      # n de contornos KDE
    kde_limiar=0.05,                   # limiar das caudas
    kde_ajuste_bw=1.2,                 # suavizao do KDE
    kde_largura_linha=1.0,             # espessura do contorno
    largura_quebra_tick=12,            # quebra de ticks (None para no quebrar)
    titulo_legenda=None                # ex.: "ms-ano"
)

# Exibe o grfico
plt.show()
```

```python
import textwrap
import matplotlib.pyplot as plt
import seaborn as sns

# --- voc j tem: media (numrico) e tabela (string "mdia  desvio") ---
# media: DataFrame com ndice = Modelo e colunas = metrics
# tabela: DataFrame com mesmo shape/ndice/colunas, com strings "xx.xx  yy.yy"

# 1) Mtricas (iguais s suas)
metrics = ['Acurcia', 'AUC', 'Recall', 'Prec.', 'F1', 'Kappa', 'MCC']

# 2) Ordenao (melhor acurcia no topo) usando a MDIA
media_sorted  = df_media.sort_values('Acurcia', ascending=False)
tabela_sorted = df_media_desvio_str.loc[media_sorted.index]

# 3) Seleo das mtricas
heatmap_data_reg = media_sorted[metrics]      # base numrica do heatmap
annot_data_reg   = tabela_sorted[metrics]     # texto "mdia  desvio"

# 4) Quebra de linha nos rtulos (Modelos)
def wrap_labels(label, width=12):
    return textwrap.fill(str(label), width)

heatmap_data_reg.index = heatmap_data_reg.index.map(lambda x: wrap_labels(x, width=20))
annot_data_reg.index   = annot_data_reg.index.map(lambda x: wrap_labels(x, width=20))

# 5) Figura
plt.figure(figsize=(19, 16))

# 6) Heatmap: cores = mdia | texto = "mdia  desvio"
sns.heatmap(
    heatmap_data_reg,
    annot=annot_data_reg,   # <- aqui est a diferena
    cmap='Blues',
    fmt="",                 # <- importante: evita o seaborn tentar formatar string como float
    linewidths=.5,
    annot_kws={"size": 24},
    cbar=False
)

# 7) Ajustes finais
plt.xticks(fontsize=22)
plt.yticks(fontsize=22, rotation=0)
plt.ylabel('')

plt.tight_layout()
plt.show()
```
